### Question2

*Donner une représentation graphique en histogramme du pourcentage de villes par région*

```python
# Read the Excel file
df = pd.read_excel("Capital.xls")


# Calculate the percentage of cities per region
region_counts = df['Région'].value_counts()
region_percentages = region_counts / len(df) * 100

# Plot the histogram
plt.bar(region_percentages.index, region_percentages)
plt.xlabel('Région')
plt.ylabel('Pourcentage de Villes')
plt.title('Pourcentage de Villes par Région')
plt.xticks(rotation=45)
plt.show()
```

### Question 1
*Faire une fonction qui permet de faire le graphe d’une variable en fonction d’une 
autre. Utiliser cette fonction pour faire le graphe de InfantDeath en fonction de Birth, 
et celui de InfantDeath en fonction de Death.*

```python
def plot_variable(df, x_variable, y_variable):
    plt.scatter(df[x_variable], df[y_variable])
    plt.xlabel(x_variable)
    plt.ylabel(y_variable)
    plt.title(f'{y_variable} vs {x_variable}')
    plt.show()

df = pd.read_table("poverty.txt", delimiter="\t", header=None, names=["Birth", "Death", "InfantDeath", "Country", "Continent"])

plot_variable(df, "Birth", "InfantDeath")
plot_variable(df, "Death", "InfantDeath")
```

### Question 2
*Proposer une autre fonction qui contiendra toutes les étapes nécessaires pour tracer
la distribution d’une variable (Birth, Death ou InfantDeath) sur les continents sous
forme d’un camembert. Utiliser ce programme sur la base poverty*

```python
def plot_pie_chart(df, variable):
    # Grouper les données par continent et calculer la moyenne de la variable choisie
    continent_means = df.groupby('Continent')[variable].mean()
    
    # Tracer le camembert
    plt.figure(figsize=(8, 6))
    plt.pie(continent_means, labels=continent_means.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribution de {variable} par continent')
    plt.axis('equal')
    plt.show()

df = pd.read_table("poverty.txt", delimiter="\t", header=None, names=["Birth", "Death", "InfantDeath", "Country", "Continent"])

plot_pie_chart(df, 'Birth')

plot_pie_chart(df, 'Death')

plot_pie_chart(df, 'InfantDeath')
```
